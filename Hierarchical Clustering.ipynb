```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
dataset = make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.6)
dataset
```

```python
points=dataset[0]
y=dataset[1]
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering 
```

```python
dendogram = sch.dendrogram(sch.linkage(points,method='ward'))
```

```python
plt.scatter(points[:,0] , points[:,1])
```

```python
plt.scatter(points[y==0,0],points[y==0,1],s=100)
plt.scatter(points[y==1,0],points[y==1,1],s=100,c='yellow')
```

```python
hc=AgglomerativeClustering(n_clusters=4, affinity='euclidean',linkage='ward')
y_hc=hc.fit_predict(points)
```

```python
print(y_hc)
```

```python
plt.scatter(points[y_hc==0,0],points[y_hc==0,1],s=100)
plt.scatter(points[y_hc==1,0],points[y_hc==1,1],s=100,c='yellow')
#plt.scatter(points[y_hc==2,0],points[y_hc==2,1],s=100,c='red')
#plt.scatter(points[y_hc==3,0],points[y_hc==3,1],s=100,c='purple')
```
